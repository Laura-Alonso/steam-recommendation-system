# ✅ Comprobación de la base de datos

```python
# Tablas almacenadas y cantidad de filas
conn = sqlite3.connect(db_path)
cur = conn.cursor()

cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cur.fetchall()]

for t in tables:
    cur.execute(f"SELECT COUNT(*) FROM {t};")
    count = cur.fetchone()[0]
    print(f"{t}: {count} filas")

conn.close()
```

```python
# Eliminar tablas de progreso temporales
conn = sqlite3.connect(db_path)
cur = conn.cursor()

tables_to_drop = ["progress"]

for t in tables_to_drop:
    cur.execute(f"DROP TABLE IF EXISTS {t};")

conn.commit()
conn.close()
```

```python
# Comprobación de la estructura de la base de datos
conn = sqlite3.connect(db_path)
cur = conn.cursor()

# Obtener todas las tablas
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cur.fetchall()]

# Listar columnas por tabla
for t in tables:
    print(f"\nTabla: {t}")
    cur.execute(f"PRAGMA table_info({t});")
    cols = cur.fetchall()
    for c in cols:
        print(f"  - {c[1]} ({c[2]})")

conn.close()
```
